# **데이터 시각화 기초**

## 1. 막대 그래프

막대 그래프는 범주별(지역별, 부문별 등)로 수치를 비교할 때 적절하다. 

pyplot 모듈의 bar 함수를 이용해서 막대그래프를 만들 수 있다.막대에 레이블을 표시할 때는 xtick 함수로 다음 예시 코드와 같이 지정한다. 기본적으로 그래프는 왼쪽 정렬이고, 중아에 놓이도록 하려면 align = 'center'로 지정한다.

```python
#  대상 데이터
x = [1,2,3]
y = [10,1,4]

# 그래프 크기 지정
plt.figure(figsize = (10, 6))
plt.bar(x,y, align='center', width = 0.5)

# 막대 그래프의 각 레이블
plt.xticks(x, ['A Class', 'B Class', 'C Class'])

# x와 y레이블 설정
plt.xlabel('Class')
plt.ylabel('Score')

# 그리드 표시
plt.grid(True)
```

**가로형 막대 그래프**

가로형 막대 그래프는 barh 함수를 사용한다. x축과 y축이 바뀌므로 좌표 이름을 다시 설정한다.

```python
#  대상 데이터
x = [1,2,3]
y = [10,1,4]

# 그래프 크기 지정
plt.figure(figsize = (10, 6))
plt.barh(x,y, align='center')

# 막대 그래프의 각 레이블
plt.yticks(x, ['A Class', 'B Class', 'C Class'])

# x와 y레이블 설정
plt.ylabel('Class')
plt.xlabel('Score')

# 그리드 표시
plt.grid(True)
```

**비교 막대 그래프 그리기**

여러 개의 막대 그래프를 그려 서로 비교할 수 있다.

```python
# 데이터 생성
y1 = np.array([30,10,40])
y2 = np.array([10,50,90])

# X축 데이터
x = np.arange(len(y1))

# 그래프 너비 
w = 0.4

# 그래프 크기 지정
plt.figure(figsize = (10,6))

# 그래프 출력. y2 막대의 너비만큼 오른쪽으로 이동시켜 그림
plt.bar(x, y1, color = 'blue', width=w, label = 'Math first', align = 'center')
plt.bar(x+w,y2, color = 'green', width = w, label = 'Math final', align = 'center')

# 좌석 위치에 범례 배치
plt.legend(loc = 'best') # 자동으로 계산하여 최적의 위치에 범례를 위치시킴

plt.xticks(x +w / 2, ['Class A', 'Class B', 'Class C'])
plt.grid(True)
```

**누적 막대 그래프**

누적 막대 그래프도 bar 함수를 사용한다. bottom 파라미터 설정에 주목한다. 

위로 누적하는 그래프는 bar의 파라미터를 bottom = <아래에 위치하는 데이터> 로 지정한다.

```python
# 데이터 생성
height1 = np.array([100,200,300,400,500])
height2 = np.array([1000,800,600,400,200])

# x축
x = np.array([1,2,3,4,5])

# 그래프 크기 지정
plt.figure(figsize = (10,6))

# 그래프 출력
p1 = plt.bar(x, height1, color = 'blue')
p2 = plt.bar(x, height2, bottom = height1, color = 'lightblue')

# 범례 표시
plt.legend((p1[0], p2[0]), ('Class 1', 'Class 2'))
```

## 2. 원 그래프

원 그래프는 전체에서 각 부분이 차지하는 비율을 볼 때 사용한다.

**일반적인 원 그래프**

원 그래프는 pie 함수를 사용해 크기와 레이블을 설정한다. 

* axis 함수로 그래프를 원형으로 지정

* autopct 파라미터에서 비율을 표시하는 형식을 지정한다. 

* explode 파라미터를 이용해 특정 범주면 전체 원에서 분리할 수도 있다.
* startangle 파라미터는 각 요소를 출력할 때 시작하는 각도를 표시하는데 시작 위치를 변경할 수 있다. (90: 왼쪽 가운데 부분이 시작위치가 되며 시계 반대 방향으로 변경하려면 양소, 시계방향은 음수로 지정한다.)
* counterclock 파라미터로 출력 방향을 지정한다. True 또는 지정하지 않는 경우 시계방향, False로 지정시 시계 반대 방향으로 출력.

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15,30,45,10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = [0,0.1,0,0]

# 그래프 크기 지정
plt.figure(figsize = (15,6))

# 그래프 출력
plt.pie(sizes, explode = explode, labels = labels, colors = colors,
        autopct = '%1.1f%%', shadow = True, startangle = 90)

# 원을 그림
plt.axis('equal')
```

# **응용: 금융 데이터 시각화**

##1. 시각화 대상 금융 데이터

다음과 같은 금융 데이터를 활용합니다.

```python
# 날짜 데이터 설정. freq='T'로 1분마다 데이터 생성
idx = pd.date_range('2015/01/01', '2015/12/31 23:59', freq='T')

# 난수 발생. 1이나 -1을 발생시킴
dn = np.random.randint(2, size = len(idx)) * 2 - 1

# 랜덤위크(무작위로 수치가 증가 또는 감소하는 데이터) 생성
# np.cumprod는 누적 곱셈 계산
# (첫 번째 원소 * 두 번째 원소 * 세 번째 원소 * ... 와 같이 차례대로 원소를 누적해서 곱한다)
rnd_walk = np.cumprod(np.exp(dn * 0.0002)) * 100

# resample('B')로 데이터를 영업일 단위로 리샘플링
# ohlc 메서드로 open, high, low, close 라는 4개의 데이터로 구분
df = pd.Series(rnd_walk, index = idx).resample('B').ohlc()
```

```python
df.plot(figsize = (15, 6), legend = 'best', grid = True)
```

## 2. 캔들 차트 생성 라이브러리

**Plotly**라이브러리가 필요하다. Plotly의 캔들 차트 기능을 이용해 아래와 같은 크래프를 출력한다. 그래프를 확대할 수도 있고 커서를 올리면 숫자가 표기된다.

```python
# plotly 모듈 임포트
from plotly.offline import init_notebook_mode, iplot
from plotly import figure_factory as FF 

# 주피터 노트북용 설정
init_notebook_mode(connected=True)

# 캔들 차트 설정
fig = FF.create_candlestick(df.open, df.high, df.low, df.close, dates = df.index)
iplot(fig)
```
