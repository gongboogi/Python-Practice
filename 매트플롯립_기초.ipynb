### 매트플롯립 사용 준비

Matplotlib은 거의 모든 그래프 기능을 pyplot.기능명 으로 제공한다.


seaborn은 매트플롯립 그래프를 더 보기 좋게 만드는 라이브러리다.

```python
# 매트플롯립과 씨본 불러오기
# 씨본은 그래프를 더 보기 좋게 만듦
import matplotlib as mpl
import seaborn as sns
import numpy as np
import numpy.random as random

# pyplot은 plt라는 이름으로 실행 가능
import matplotlib.pyplot as plt

# 주피터 노트북에서 그래프를 표시하기 위해 필요한 매직 명령어
%matplotlib inline
```

### 산점도

산점도는 2개의 데이터 값을 x - y 좌표에 점으로 나타낸 그래프다. 

plt.plot(x, y, 'o') 구문으로 그래프를 그릴 수 있다. 마지막 인수는 그래프의 형태를 지정하는 값으로 'o'는 점을 찍는다는 뜻이다.

```python
# 산점도
# 시드 값 설정
random.seed(0)

# x축 데이터
x = np.random.randn(30)

# y축 데이터
y = np.sin(x) + random.randn(30)

# 그래프 크기 지정
plt.figure(figsize = (15,6))

# 그래프 생성
plt.plot(x, y, 'o')

# scatter로도 산점도 생성 가능
plt.scatter(x,y)

# 타이틀
plt.title('Title Name')

# x 좌표 이름
plt.xlabel('X')

# y 좌표 이름
plt.ylabel('Y')

# grid(그래프 안의 가로선과 세로선 격자) 표시
plt.grid(True)
```

데이터가 연속된 값이면 plot으로 그려지는 그래프는 점이 아니라 곡선으로 보인다.

시계열 데이터와 같은 연속적인 곡선을 그린다.

```python
# 연속형 공선

# 시드 값 지정
np.random.seed(0)

# 데이터 범위
data_x = np.arange(1000)

# 난수 발생과 누적 합계
random_data_y = random.randn(1000).cumsum()

# 그래프 크기 지정
plt.figure(figsize = (15, 6))

# label과 legend로 레이블을 표시할 수 있다.
plt.plot(data_x, random_data_y, label = 'Label')
plt.legend()

plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

### 그래프 분할

subplot을 이용하면 그래플르 여러 개로 분할해 나타낼 수 있다. 

다음 코드는 2행 1열 그래프를 생성하고 번호를 붙여 표시하는 예다.

```python
# 그래프 크기 지정
plt.figure(figsize = (15,6))

# 2행 1열 그래프의 첫 번째
plt.subplot(2,1,1)

x = np.linspace(-10, 10, 100)  # -10부터 10까지의 수를 100개로 분할한 숫자 리스트 추출
plt.plot(x, np.sin(x))

# 2행 1열 그래프 두 번째
plt.subplot(2,1,2)

y = np.linspace(-10, 10, 100)
plt.plot(y, np.sin(2**y))

plt.grid(True)
```



## 함수 그래프 그리기

```python
# 함수 정의(사이파이 기초에서 사용한 2차원 함수 예와 동일)
def my_function(x) :
  return x ** 2 + 2 * x + 1

x = np.arange(-10, 10)

plt.figure(figsize = (15,6))
plt.plot(x, my_function(x))
plt.grid(True)
```

## 히스토그램

히스토그램은 각 값의 빈도(도수, freuquency)를 표시한다. 데이터의 전체적인 형태를 볼 때 사용하는 그래프다.

hist 메서드를 사용하며, 괄호 안에 지정하는 파라미터는 (대상데이터, 막대수, 범위)다.

```python
# 히스토그램
# 시드지정
random.seed(0)

# 그래프 크기 지정
plt.figure(figsize = (15,6))

# 히스토그램 생성
plt.hist(np.random.randn(10 ** 5) * 10 + 50, bins = 60, range = (20, 80))

plt.grid(True)
```

### 3. 0부터 1사이의 값을 갖는 균등분포 난수를 1000개씩 두 번 발생시킨 후 각 데이터의 히스토그램을 나타내세요.



```python
import math

def uni_hist(N):
  # 균등분포 난수 발생
  x = np.random.uniform(0.0, 1.0, N)
  y = np.random.uniform(0.0, 1.0,N)

  plt.subplot(2,1,1)
  plt.hist(x)
  plt.title("No1: histogram")

  plt.subplot(2,1,2)
  plt.hist(y)
  plt.title("No2: histogram")

  plt.grid(True)

  # 그래프 제목 겹침 방지 자동 레이아웃 설정
  plt.tight_layout()

uni_hist(1000)
```
